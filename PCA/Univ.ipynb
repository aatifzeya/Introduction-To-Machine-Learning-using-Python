```python
import pandas as pd 
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale 
```

```python
uni = pd.read_csv("Univ.csv")
uni.describe()
uni.head()
```

```python
# Considering only numerical data 
uni.data = uni.iloc[:,1:]
uni.data.head()
# Converting into numpy array
UNI = uni.data.values
UNI
```

```python
# Normalizing the numerical data 
uni_normal = scale(UNI)
uni_normal
```

```python
import seaborn as sns
sns.pairplot(uni.data)
```

```python
pca = PCA(n_components = 6)
pca_values = pca.fit_transform(uni_normal)#pca_values
data=pd.DataFrame(pca_values)
data
```

```python
#pca = PCA(n_components = 6)
#pca_values = pca.fit_transform(uni_normal)
```

```python
# The amount of variance that each PCA explains is 
pca.explained_variance_ratio_
```

```python
var=(pca.explained_variance_ratio_)*100
var
```

```python
# Cumulative variance 
cumvar = np.cumsum(np.round(var,decimals = 1))
cumvar
```

```python
#ploting PCAs in pairplot.we can see that correlation between the PCs is close to zero
sns.pairplot(pd.DataFrame(pca_values))
```

```python
pd.DataFrame(pca_values).corr()
```

```python
pca.components_
```

```python
# Variance plot for PCA components obtained 
plt.plot(cumvar,color="Green")
```

```python
finalDf = pd.concat([pd.DataFrame(pca_values[:,0:2],columns=['pc1','pc2']), uni[['Univ']]], axis = 1)
finalDf.head()
```
